```python
import networkx as nx

# Read the graph from the text file
def read_graph(file_path):
    G = nx.DiGraph()
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith("#"):
                continue
            else:
                edge = line.strip().split()
                if len(edge) == 2:
                    from_node, to_node = map(int, edge)
                    G.add_edge(from_node, to_node)
    return G

# Function to perform PageRank analysis
def page_rank_analysis(graph):
    # Calculate PageRank
    pagerank_scores = nx.pagerank(graph)
    # Sort nodes by PageRank score
    sorted_nodes = sorted(pagerank_scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_nodes

# Path to the text file containing the graph data
file_path = 'web-Stanford.txt'

# Read the graph from the text file
graph = read_graph(file_path)

# Perform PageRank analysis
pagerank_result = page_rank_analysis(graph)

# Print the top nodes and their PageRank scores
top_nodes_count = 100
print(f"Top {top_nodes_count} nodes by PageRank:")
for i in range(top_nodes_count):
    node, score = pagerank_result[i]
    print(f"Node {node}: PageRank Score = {score}")
```
